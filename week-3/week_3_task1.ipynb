```python
def data_cleanning():
    #dealling with null values of Age column, by replacing all null values with mean of the age.
    age_mean=np.mean(data['Age']).round(decimals=1)
    data['Age']=data['Age'].fillna(age_mean)

    #salary column
    data.insert(2,column='Start_Salary',value=0)
    data.insert(3,column="End_Salary",value=0)
    data['Salary']=data['Salary'].str.replace("$","")
    data['Salary']=data['Salary'].str.replace("k","000")

    index=0
    for i in data['Salary']:
        j=i.split('-')
        data['Start_Salary'][index]=int(j[0])
        data['End_Salary'][index]=int(j[1])
        index+=1
    data.pop('Salary')

    #Rating column
    index1=0
    for i in data['Rating']:
        if(i<0):
            data['Rating'][index1]=0
        # print(i)
        index1+=1
    #filling null value with zero
    data['Rating'] = data['Rating'].fillna(0)

    #Country column
    data['Location'] = data['Location'].str.replace(","," ")
    index=0
    for i in data['Location']:
        remove = i.split()[-1]
        data['Location'][index] = data['Location'][index].rstrip(remove)
        index+=1

    #Established
    data['Established'] = data['Established'].replace(-1,"NA")

    #Easy Apply
    data['Easy Apply'] = data['Easy Apply'].str.replace("-1","FALSE")

data_cleanning()
```
